```python
import pandas as pd
import numpy as np
import datetime
```

```python
cc_data=pd.read_csv('cc_data.csv',encoding='latin-1')
loyalty_data=pd.read_csv('loyalty_data.csv',encoding='latin-1')
cars_data=pd.read_csv('car-assignments.csv',encoding='latin-1')
gps_data=pd.read_csv('gps.csv',encoding='latin-1')
```

# Employees spending per business

```python
cars_data['fullName']=cars_data['FirstName']+" "+cars_data['LastName']
cars_data=cars_data[['CarID','fullName']]
cars_data=cars_data.fillna(0)
cars_data.head()
```

```python
cars_data.tail()
```

```python
cc_date=pd.to_datetime(cc_data.timestamp).dt.date
cc_data['date']=cc_date
cc_data['fullName']=cc_data['FirstName']+" "+cc_data['LastName']
cc_data=cc_data[['timestamp','date','location','price','fullName']]
cc_data.head()
```

```python
#find employee's carID
newcc_data= cars_data.merge(cc_data,how='outer')
newcc_data.head()
```

```python
cc_sum= newcc_data.groupby(['fullName','CarID', 'location']).sum()
cc_sum.head()
```

```python
cc_sum.to_csv('NEW_cc_sum.csv')
```
